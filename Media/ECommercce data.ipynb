```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("oleksiymaliovanyy/call-center-transcripts-dataset")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

# List files in the downloaded directory
print("Files in dataset directory:")
for root, dirs, files in os.walk(path):
    for file in files:
        print(os.path.join(root, file))

# Assuming the main dataset is 'call_recordings.csv' based on the file listing
dataset_file = os.path.join(path, 'call_recordings.csv')

if os.path.exists(dataset_file):
    print(f"\nLoading dataset from: {dataset_file}")
    df = pd.read_csv(dataset_file)

    print("\nDataset loaded successfully. Here's the head of the DataFrame:")
    display(df.head())
    print("\nDataset info:")
    df.info()
else:
    print(f"\nError: Expected dataset file not found at {dataset_file}. Please inspect the directory structure manually.")
```
