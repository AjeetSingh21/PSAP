```python
import os

# Define common audio file extensions
audio_extensions = ('.wav', '.mp3', '.m4a', '.flac', '.aac', '.ogg', '.wma')

# List to store full paths of audio files
audio_files = []

# Iterate through the directory contents
for filename in os.listdir(path):
    if filename.lower().endswith(audio_extensions):
        full_path = os.path.join(path, filename)
        audio_files.append(full_path)

# Print the identified audio files
if audio_files:
    print(f"Found {len(audio_files)} audio files:")
    for audio_file in audio_files:
        print(audio_file)
else:
    print("No audio files found in the specified directory.")
```

```python
import os

# Define common audio file extensions
audio_extensions = ('.wav', '.mp3', '.m4a', '.flac', '.aac', '.ogg', '.wma')

# List to store full paths of audio files
audio_files = []

# Iterate through the directory and its subdirectories
for root, dirs, files in os.walk(path):
    for filename in files:
        if filename.lower().endswith(audio_extensions):
            full_path = os.path.join(root, filename)
            audio_files.append(full_path)

# Print the identified audio files
if audio_files:
    print(f"Found {len(audio_files)} audio files:")
    for audio_file in audio_files:
        print(audio_file)
else:
    print("No audio files found in the specified directory or its subdirectories.")
```

```python
import speech_recognition as sr
from pydub import AudioSegment
import os

# List to store filename and transcription
transcriptions = []

# Create a Recognizer object
r = sr.Recognizer()

# Iterate through each audio file
for i, audio_file_path in enumerate(audio_files):
    print(f"Processing file {i+1}/{len(audio_files)}: {os.path.basename(audio_file_path)}")

    temp_wav_file = None
    try:
        # Check if the file is an MP3 and convert to WAV
        if audio_file_path.lower().endswith('.mp3'):
            audio = AudioSegment.from_mp3(audio_file_path)
            temp_wav_file = "temp_audio.wav"
            audio.export(temp_wav_file, format="wav")
            source_file = temp_wav_file
        else:
            # For other formats, use the original file path
            source_file = audio_file_path

        with sr.AudioFile(source_file) as source:
            # Adjust for ambient noise
            r.adjust_for_ambient_noise(source)
            # Record the audio data
            audio_data = r.record(source)

            # Attempt to transcribe using Google Speech Recognition
            try:
                text = r.recognize_google(audio_data)
                transcriptions.append({'filename': os.path.basename(audio_file_path), 'transcription': text})
                print(f"  Transcription: {text[:70]}...") # Print first 70 chars
            except sr.UnknownValueError:
                transcriptions.append({'filename': os.path.basename(audio_file_path), 'transcription': 'Could not understand audio'})
                print("  Transcription: Could not understand audio")
            except sr.RequestError as e:
                transcriptions.append({'filename': os.path.basename(audio_file_path), 'transcription': f"Transcription service error: {e}"})
                print(f"  Transcription service error: {e}")

    except Exception as e:
        transcriptions.append({'filename': os.path.basename(audio_file_path), 'transcription': f"Error processing file: {e}"})
        print(f"  Error processing file: {e}")
    finally:
        # Clean up temporary WAV file if it was created
        if temp_wav_file and os.path.exists(temp_wav_file):
            os.remove(temp_wav_file)

print(f"\nCompleted transcription for {len(transcriptions)} audio files.")
```

```python
import pandas as pd

# Convert the list of dictionaries to a pandas DataFrame
transcriptions_df = pd.DataFrame(transcriptions)

# Define the output CSV file path
output_csv_path = 'audio_transcriptions.csv'

# Save the DataFrame to a CSV file
transcriptions_df.to_csv(output_csv_path, index=False)

print(f"Transcriptions saved to {output_csv_path}")
print("First 5 rows of the DataFrame:")
print(transcriptions_df.head())
```
