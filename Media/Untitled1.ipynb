## Generate AI Voice Greeting using Azure Text-to-Speech

To use the Azure Text-to-Speech API, you'll need an Azure subscription and a Text-to-Speech resource. Once you have that, you can obtain an API key and the region for your resource.

In Colab, add your Azure Text-to-Speech API key to the secrets manager under the "🔑" in the left panel. Give it the name `AZURE_TTS_KEY`. Also, add your Azure Text-to-Speech region (e.g., `eastus`, `westus2`) as `AZURE_TTS_REGION`.

```python
# Import necessary libraries
import requests
from google.colab import userdata
from IPython.display import Audio
import os

# Retrieve Azure credentials from Colab secrets
AZURE_TTS_KEY = userdata.get('AZURE_TTS_KEY')
AZURE_TTS_REGION = userdata.get('AZURE_TTS_REGION')
```

```python
# Define the voice and the greeting message
voice_name = "en-IN-AartiNeural"
greeting_text = "Hello, thank you for calling. Please wait while I connect you to our appointment scheduling service."

# Azure Text-to-Speech endpoint
# Construct the endpoint URL using the region
endpoint = f"https://{AZURE_TTS_REGION}.tts.speech.microsoft.com/cognitiveservices/v1"

# SSML (Speech Synthesis Markup Language) payload
# This allows specifying the voice and other speech characteristics
ssml_body = f"""<speak version='1.0' xml:lang='en-US'>
    <voice name='{voice_name}'>
        {greeting_text}
    </voice>
</speak>"""

# Request headers
headers = {
    'X-Microsoft-OutputFormat': 'audio-16khz-32kbitrate-mono-mp3',
    'Content-Type': 'application/ssml+xml',
    'Host': f'{AZURE_TTS_REGION}.tts.speech.microsoft.com',
    'Ocp-Apim-Subscription-Key': AZURE_TTS_KEY,
    'User-Agent': 'ColabPythonClient'
}
```

```python
# Make the request to the Azure Text-to-Speech API
try:
    response = requests.post(endpoint, headers=headers, data=ssml_body.encode('utf-8'))
    response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)

    # Save the audio content to a file
    output_filename = "greeting_voice.mp3"
    with open(output_filename, 'wb') as audio_file:
        audio_file.write(response.content)

    print(f"Audio successfully generated and saved as '{output_filename}'")

    # Play the audio directly in Colab
    display(Audio(output_filename))

except requests.exceptions.HTTPError as err:
    print(f"HTTP error occurred: {err}")
    print(f"Response content: {response.text}")
except requests.exceptions.RequestException as err:
    print(f"An error occurred during the request: {err}")
except Exception as err:
    print(f"An unexpected error occurred: {err}")
```

If you want to modify the greeting message or try a different voice (ensure it's compatible with Azure's Text-to-Speech service and your chosen region), you can edit the `greeting_text` and `voice_name` variables in the previous cells and re-run them.
